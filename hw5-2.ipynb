```python
# networkx only
import networkx as nx

def deBruijn(l):
    """ Takes in input a list of strings l of lenght k and
        returns a deBruijn multi-edge directed graph """
    g = nx.MultiDiGraph()      # we need directed multigraph because we can have multiple edges between two nodes
    ## YOUR CODE HERE TO BUILD THE GRAPH
    for s in l:
        prefix = s[:-1]
        suffix = s[1:]
        g.add_edge(prefix, suffix)
    return g

with open("hw5_test13.txt", "r") as f: # open the file
    w = [line.rstrip() for line in f] # read the words
g = deBruijn(w)                       # build the graph

#for list in w:
#    print(list)  this would be to print a list of the k-mers
adjacency = {}
for u, v in g.edges():
    if u not in adjacency:  #was failing because was not accounting for repeated strings
        adjacency[u] = [v]
    else:
        adjacency[u].append(v)  #if its already present, we just add the repetition to where it repeats


## YOUR CODE HERE 
## print the edges
for prefix in sorted(adjacency.keys()): #expected output is also sorted, so we need to sort it
    suffixes = ",".join(adjacency[prefix]) 
    print(f"{prefix} -> {suffixes}") #print the connections sorted
```
